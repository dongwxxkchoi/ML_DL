```python
import numpy as np

class Perceptron():
    # 4가지 경우 0,0 0,1 1,0 1,1에 대해 AND연산을 학습시키는 것이 목표다
    def __init__(self, thresholds=0.0, eta=0.01, n_iter=10): # thresholds -> 임곗값
        # eta -> estimated time of arrival / n_iter = 반복 횟수
        self.thresholds = thresholds
        self.eta = eta
        self.n_iter = n_iter
    
    def fit(self, X, y): 
        self.w_ = np.zeros(1+X.shape[1]) # 1을 더하는 것은 x0는 편향치라 w0는 사용 x
        # shape[1] means num of rows (열단위로 하나봄) / 보통 가중치는 0이기 때문에, 0으로 맞춰 놓음
        self.errors_ = []
        
        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y): # iterator화 하기 -> X와 y의 차원이 같아야 제대로
                print("---------------------------------")
                print("xi: ", xi, "target: ", target)
                predict_val = self.predict(xi)
                update = self.eta * (target-predict_val) # 오차를 구해 가중치 업데이트 (target-predict_val)
                # 만약 predict_val이 target과 같다, 즉 예측 성공적으로 하면 -> update 안 함
                print("predict_val: ", predict_val, "update: ", update)
                self.w_[1:] += update * xi
                self.w_[0] += update
                errors += int(update!=0.0)
                print("after update")
                print("w: ", self.w_, "errors: ", errors, "(0이면 error 없었음, 1이면 있었음, 이 error는 누적치)")
                print("---------------------------------")
            self.errors_.append(errors)
            # turn 마다 error가 몇 번 일어났는지를 기록하기 위한 리스트
            print("this turn's 가중치: ", self.w_)
        
        return self

    def net_input(self, X):
        result = np.dot(X, self.w_[1:]) + self.w_[0]
        print("가중치:", self.w_, "net_input result: ", result)
        # np.dot -> 행렬곱  /  X는 1x2, w_[:1]는 2x1 -> scalar value가 산출됨
        return result # 행렬곱에 가중치가 더해져 산출됨
    
    def predict(self, X): # predict를 통해 실제값을 예측함
        return np.where(self.net_input(X) > self.thresholds, 1, -1)
        # 조건을 만족하는 self.net_input(X)의 index에 대해 True면 1, False면 -1로 바꿈
        # 임곗값(0)보다 net_input이 커야 1, 작으면 -1
        
```

```python
import numpy as np

class AdalineGD():
    def __init__(self, eta=0.01, n_iter=50):
        self.eta = eta
        self.n_iter = n_iter
    
    def fit(self, X, y): # X를 y에 맞추는
        self.w_ = np.zeros(1+X.shape[1])
        self.cost_ = [] 
        # 비용함수 J(w)가 어떤 값으로 수렴하는지 확인하기 위해
        # cost? -> 매 반복마다 계산되는 비용함수 값을 저장하기 위한 변수
        # J(w)는 오차를 실제와 data*가중치를 뺀것을 제곱해서 구함
        
        for i in range(self.n_iter):
            output = self.net_input(X) # 행렬 내적
            errors = (y - output) # 오차값
            self.w_[1:] += self.eta * X.T.dot(errors) 
            # 가중치 업데이트
            # self.eta -> learning rate
            # X.T와 errors의 행렬곱이 시그마에 해당
            self.w_[0] += self.eta * errors.sum() 
            # 편향 더하는 부분
            # errors의 합과 eta 곱해 편향에 반영
            cost = (errors**2).sum() / 2
            # J(w) 결과 -> errors는 y - 시그마(wjxj)를 뜻함
            self.cost_.append(cost)
        
        return self

    def net_input(self, X):
        return np.dot(X, self.w_[1:] + self.w_[0])

    def predict(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)
    
```

```python
%matplotlib inline
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style

if __name__ == "__main__":
    df = pd.read_csv('iris.data', header=None)
    y = df.iloc[0:100, 4].values
    y = np.where(y=='Iris-setosa', -1, 1)
    X = df.iloc[0:100, [0, 2]].values
    
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
    
    adal = AdalineGD(eta=0.01, n_iter=10).fit(X, y)
    ax[0].plot(range(1, len(adal.cost_) + 1), np.log10(adal.cost_), marker='o')
    ax[0].set_xlabel('Epochs')
    ax[0].set_ylabel('log(SQE)')
    ax[0].set_title('Adaline - Learning rate 0.01')
    
    adal2 = AdalineGD(eta=0.0001, n_iter=10).fit(X, y)
    ax[1].plot(range(1, len(adal2.cost_) + 1), np.log10(adal2.cost_), marker='o')
    ax[1].set_xlabel('Epochs')
    ax[1].set_ylabel('log(SQE)')
    ax[1].set_title('Adaline - Learning rate 0.0001')
    
    plt.subplots_adjust(left=0.1, bottom=0.2, right=0.95, top=0.9, wspace=0.5, hspace=0.5)
    
    plt.show()
```

### 이 방법을 경사하강법 (Batch Gradient Descent)을 이용한 아달라인이라 부릅니다
##### -> 안 좋은 점은, 데이터의 개수가 많아지고, 종류가 다양하면 효율적인 수행이 불가하다
이를 해결하는 방법으로 확률적 경사하강법을 이용하는 아달라인을 알아보자
-> 다른 점은, 모든 트레이닝 데이터의 결괏값-예측값을 다 더하는 것이 아닌, i번째 트레이닝 데이터에 대해서만 계산한 값을 이용

```python
import numpy as np
from numpy.random import seed

class AdalineSGD():
    def __init__(self, eta=0.01, n_iter=10, shuffle=True, random_state=None):
        self.eta = eta
        self.n_iter = n_iter
        self.shuffle = shuffle
        
        if random_state:
            seed(random_state)
    
    def fit(self, X, y):
        self.w_ = np.zeros(1+X.shape[1])
        self.cost_ = []
        
        for i in range(self.n_iter):
            if self.shuffle:
                X, y = self._shuffle(X, y)
                
            cost = []
            for xi, target in zip(X, y):
                output = self.net_input(xi)
                error = target - output
                #self.w_[1:] += self.eta * X.T.dot(errors) 원래
                #self.w_[0] += self.eta * errors.sum() 원래
                self.w_[1:] += self.eta * xi.dot(error) # -> X가 아닌 xi를, errors가 아닌 error를 dot
                self.w_[0] += self.eta * error # -> errors.sum()이 아닌 error
                cost.append(0.5 * error**2) # -> 여기 역시 sum이 아닌 그냥 error
                
            avg_cost = sum(cost)/len(y)
            self.cost_.append(avg_cost)
            
        return self
    
    def _shuffle(self, X, y):
        r = np.random.permutation(len(y))
        return X[r], y[r]
    
    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]
    
    def predict(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

```python
%matplotlib inline
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style

style.use('seaborn-talk')

if __name__ == "__main__":
    df = pd.read_csv('iris.data', header=None)
    y = df.iloc[0:100, 4].values
    y = np.where(y=='iris-setosa', -1, 1)
    X = df.iloc[0:100, [0, 2]].values
    
    X_std = np.copy(X)
    X_std[:, 0] = (X[:, 0] - X[:, 0].mean()) / X[:, 0].std()
    X_std[:, 1] = (X[:, 1] - X[:, 1].mean()) / X[:, 1].std()
    
    adal = AdalineSGD(eta=0.01, n_iter=15, random_state=1).fit(X_std, y)
    plt.plot(range(1, len(adal.cost_) + 1), adal.cost_, marker='o')
    plt.xlabel('Epochs')
    plt.ylabel('Average Cost')
    plt.title('Adaline Stochastic GD - Learning rate 0.01')
    plt.show()
    
```
